#### Q - 1

Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#### lets first connect to the server

driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
name_s=[]
artist_s=[]
```

```python
# scrapping name_s from the given page
names_tags=driver.find_elements(By.XPATH,'//a[@class="mw-redirect"]')
for i in names_tags[0:10]:
    names=i.text
    name_s.append(names)
    
# scrapping ARTIST_S from the given page

artist_tags=driver.find_elements(By.XPATH,'//td[@align="right"]')
for a in names_tags[0:10]:
    artist=a.text
    artist_s.append(artist)
    
```

```python
print(len(name_s),len(artist_s))
```

```python
import pandas as pd
df=pd.DataFrame({'Names':name_s,'Artists':artist_s})
df
```

```python
driver.close()
```

#### Q - 2

Scrape the details team India’s internationalfixtures from bcci.tv. 

Url = https://www.bcci.tv/.

You need to find following details:

A) Match title (I.e. 1stODI)

B) Series

C) Place

D) Date

E) Time


```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
#### lets first connect to the server

driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# opening the automated page on automated chrome browser
driver.get("https://www.bcci.tv/")   # opening the automated page on automated chrome browser
import time
```

```python
match_title=[]
place=[]
date=[]
time=[]
```

```python
# scrapping match_title from the given page
names_tags=driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
for i in names_tags[0:10]:
    names=i.text
    match_title.append(names)
    
# scrapping place from the given page
place_tags=driver.find_elements(By.XPATH,'//span[@class="ng-binding ng-scope"]')
for i in place_tags[0:10]:
    plac=i.text
    place.append(plac)
    
# scrapping date from the given page
date_tags=driver.find_elements(By.XPATH,'//h5[@class="ng-binding"]')
for i in date_tags[0:10]:
    dat=i.text
    date.append(dat)
    
#  scrapping time from the given page
time_tags=driver.find_elements(By.XPATH,'//h5[@class="text-right ng-binding"]')
for i in time_tags[0:10]:
    tim=i.text
    time.append(tim)
```

```python
print(len(match_title),len(place),len(date),len(time))
```

```python
import pandas as pd
df=pd.DataFrame({'Match_title':match_title,'Place':place,'Date':date,'Time':time})
df
```

```python
driver.close()
```

#### Q - 3 

Scrape the details of State-wise GDP ofIndia fromstatisticstime.com.

Url = http://statisticstimes.com/
    
You have to find following details:
    
A) Rank

B) State

C) GSDP(18-19)- at current prices

D) GSDP(19-20)- at current prices

E) Share(18-19)

F) GDP($ billion)


Note: - From statisticstimes home page you have to reach to economy page through code.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By

import time
```

```python
#### lets first connect to the server

driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
# opening the automated page on automated chrome browser
driver.get("https://www.statisticstimes.com/")   # opening the automated page on automated chrome browser
import time
```

```python
Rank=[]
State=[]
GSDP_1819=[]
GSDPP_1920=[]
Share_1819=[]
GDP_billion=[]
```

```python
# scrapping rank from the given page
rank_tags=driver.find_elements(By.XPATH,'//td[@class="data1"]')
for i in rank_tags[0:33]:
    rank=i.text
    Rank.append(rank)
    
# scrapping state from the given page
state_tags=driver.find_elements(By.XPATH,'//td[@class="name"]')
for i in state_tags[0:33]:
    state=i.text
    State.append(state)
    
# scrapping GSDP_1819 from the given page
gsdp_1819_tags=driver.find_elements(By.XPATH,'//td[@class="data sorting_1"]')
for i in gsdp_1819_tags[0:33]:
    gsdp_1819=i.text
    GSDP_1819.append(gsdp_1819)
    
#  scrapping gsDPP_1920 from the given page
gSDPP_1920_tags=driver.find_elements(By.XPATH,'//td[@class="data"]')
for i in gSDPP_1920_tags[0:33]:
    gSDPP_1920=i.text
    GSDPP_1920.append(gSDPP_1920)
    
# scrapping Share_1819 from the given page
share_1819_tags=driver.find_elements(By.XPATH,'//td[@class="data"]')
for i in share_1819_tags[0:33]:
    share_1819=i.text
    Share_1819.append(share1819)
    
#  scrapping GDP_billion from the given page
gDP_billion_tags=driver.find_elements(By.XPATH,'//td[@class="data"]')
for i in gDP_billion_tags[0:33]:
    gDP_billion=i.text
    GDP_billion.append(gDP_billion)
```

```python
print(len(Rank),len(State),len(GSDP_1819),len(GSDPP_1920),len(Share_1819),len(GDP_billion))
```

```python
import pandas as pd
df=pd.DataFrame({'rank':Rank,'state':State,'gSDP_1819':GSDP_1819,'gSDPP_1920':GSDPP_1920,'share_1819':Share_1819,'gDP_billion':GDP_billion})
df
```

```python
driver.close()
```
